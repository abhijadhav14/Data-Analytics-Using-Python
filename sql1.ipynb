```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("SparkSQLExample").getOrCreate()
```

```python
data = [
    (1,"Logan Paul",60), (2,"KSI",75), (3,"IShowSpeed",53),
    (4,"Lewis Hamilton",40), (5,"Max Verstappen",38), (6,"Fernando Alanzo",55),
    (7,"Nico Hulkenberg",42), (8,"Oscar Paistri",33), (9,"Brad Pitt",46), (10,"Tom Hanks",49)
]
columns = ["Id", "Name", "Age"]
```

```python
df = spark.createDataFrame(data, columns)
print("Original DataFrame:")
df.show()
```

```python
# DataFrame as a temporary SQL view
df.createOrReplaceTempView("Actors")
```

```python
# SQL query
print("Actors older than 50:")
result = spark.sql("SELECT * FROM Actors WHERE age > 50")
result.show()
```

```python
# SQL query
print("Average age:")
avg_age = spark.sql("SELECT AVG(age) AS average_age FROM Actors")
avg_age.show()
```

```python
spark.stop()
```
