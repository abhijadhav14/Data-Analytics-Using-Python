```python
from pyspark import SparkContext
```

```python
file_path = "/content/drive/MyDrive/DAP/accesslog.txt"
```

```python
try:
  sc.stop()

except:
  pass

sc = SparkContext("local[*]","Experiment4")
rdd = sc.textFile(file_path)
for line in rdd.take(5):
  print(line)
```

```python
import re
```

```python
log_patern = re.compile(r'(\S+) (\S+) (\S+) \[(.*?)\] "(.*?)" (\d{3}) (\S+)')
```

```python
def parse_log(line):
  match = log_patern.match(line)

  if match:
    return {
        "ip": match.group(1),
        "time'": match.group(4),
        "request": match.group(5),
        "status": match.group(6),
        "size": match.group(7)
    }

  return none
```

```python
parsed_rdd = rdd.map(parse_log).filter(lambda x: x is not None)
total_requests = parsed_rdd.count()
print(total_requests)
```

```python
ip_counts = (parsed_rdd.map(lambda x: (x["ip"], 1)).reduceByKey(lambda a, b: a + b).sortBy(lambda x: x[1], ascending=False))
```

```python
for ip, count in ip_counts.take(10):
  print(f"IP: {ip}, Count: {count}")
```
