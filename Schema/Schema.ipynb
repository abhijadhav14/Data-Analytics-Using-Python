```python
# Defining Schema

from pyspark.sql import SparkSession
from pyspark.sql.types import *

spark = SparkSession.builder.getOrCreate()

myschema = StructType([
    StructField("Date", IntegerType(), True),
    StructField("Items_on_Sale", DoubleType(), True),
    StructField("Discount", DoubleType(), True),
    StructField("Sales", DoubleType(), True)
])

df = spark.read.schema(myschema).csv("sample_sales_data.csv")
df.show()
df.printSchema()
```

```python
# Datatypes

# Numeric:
#   ByteType
#   ShortType
#   IntegerType
#   LongType
#   FloatType
#   DoubleType
#   DecimalType

# String/Binary:
#   StringType
#   BinaryType
#   VarcharType
#   CharType

# Datetime/Boolean:
#   DateType
#   TimestampType
#   BooleanType
#   TimeType
```

```python
# Write a code snippet to demonstrate innferred Schema and User Defined Schema(Use two Different CSV files)

# 1) Inferred schema (reads a different file with header)
df_inferred = spark.read.option("header", True).option("inferSchema", True).csv("file1.csv")
print("Inferred Schema:")
df_inferred.printSchema()
df_inferred.show()
```

```python
# 2) User-defined schema (reads file2.csv without header)
user_schema = StructType([
    StructField("Date", DateType(), True),
    StructField("Items_on_Sale", DoubleType(), True),
    StructField("Discount", IntegerType(), True),
    StructField("Sales", DoubleType(), True)
])

df_user_defined = spark.read.schema(user_schema).csv("file2.csv")
print("User Defined Schema:")
df_user_defined.printSchema()
df_user_defined.show()
```
