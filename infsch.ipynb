```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('InferredSchema1').getOrCreate()
```

```python
data = [("Chennai", 34), ("Delhi", 45),("Bangalore", 29), ("Mumbai", 42)]
```

```python
columns = ["City", "Temperature"]
```

```python
df = spark.createDataFrame(data, columns)
```

```python
df.show()
df.printSchema()
```
